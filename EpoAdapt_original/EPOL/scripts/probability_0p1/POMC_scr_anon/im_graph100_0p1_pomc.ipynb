**---------Budget 100 Results----------**

```
import re
import numpy as np

# The result strings
results = [
    'value = 59.93',
    'value = 60.064',
    'value = 59.874'
]

# Extract times using regex
total_accumulated_values = []

for line in results:
    match = re.search(r'value =\s*([\d.]+)', line)
    if match:
        total_accumulated_values.append(float(match.group(1)))
    else:
        print(f"Warning: No match found in line: {line}")

# Compute mean, std, and CV
def stats(name, values):
    avg = np.mean(values)
    std = np.std(values, ddof=1)  # Sample std dev
    cv = (std / avg) * 100 if avg != 0 else float('nan')  # CV in percentage
    print(f"{name} - Mean: {avg:.4f} sec, Std Dev: {std:.4f} sec, CV: {cv:.2f}%")

stats("Total Value", total_accumulated_values)
```

**---------Time Analysis---------**

```
import re
import numpy as np

# The result strings
results = [
    'Elapsed time: 653.6854',
    'Elapsed time: 603.2059',
    'Elapsed time: 639.3115'
]

# Extract times using regex
total_accumulated_values = []

for line in results:
    match = re.search(r'Elapsed time:\s*([\d.]+)', line)
    if match:
        total_accumulated_values.append(float(match.group(1)))
    else:
        print(f"Warning: No match found in line: {line}")

# Compute mean, std, and CV
def stats(name, values):
    avg = np.mean(values)
    std = np.std(values, ddof=1)  # Sample std dev
    cv = (std / avg) * 100 if avg != 0 else float('nan')  # CV in percentage
    print(f"{name} - Mean: {avg:.4f} sec, Std Dev: {std:.4f} sec, CV: {cv:.2f}%")

stats("Total Time", total_accumulated_values)
```
