```python
# Xem cấu trúc của các file dataset
import os

# Đọc và phân tích file mẫu
def analyze_dataset(filepath):
    """Phân tích cấu trúc dataset"""
    with open(filepath, 'r') as f:
        lines = [line.strip() for line in f.readlines()]
    
    edges = [line.split() for line in lines if line]
    nodes = set()
    for edge in edges:
        nodes.add(int(edge[0]))
        nodes.add(int(edge[1]))
    
    print(f"File: {os.path.basename(filepath)}")
    print(f"  Nodes: {len(nodes)} (từ {min(nodes)} đến {max(nodes)})")
    print(f"  Edges: {len(edges)}")
    print(f"  Format: edge list (source target)")
    print(f"  Ví dụ 5 dòng đầu: {lines[:5]}")
    return len(nodes), len(edges)

# Phân tích các file trong EPOL_modified
epol_dir = "/home/kongwoang/data/MSO_Lab/SubsetSelection/EPOL_modified/outdegree"
sample_files = ["graph100-01.txt", "graph200-01.txt", "insecta.txt"]

print("=" * 70)
print("CẤU TRÚC DATASET - Directed Graph Edge Lists")
print("=" * 70)
print("\nCác file thể hiện đồ thị có hướng dạng edge list:")
print("- Mỗi dòng: <node_nguồn> <node_đích>")
print("- Dùng cho bài toán Influence Maximization")
print("- Đồ thị mô tả mạng lưới lan truyền ảnh hưởng\n")

for fname in sample_files:
    fpath = os.path.join(epol_dir, fname)
    if os.path.exists(fpath):
        analyze_dataset(fpath)
        print()

print("\nFile _eps.txt: Có thể là graph với epsilon values cho uncertainty")
```
